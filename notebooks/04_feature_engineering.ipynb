# **Feature Engineering**

### Importing Libraries

```python
import pandas as pd
import numpy as np
```

## Importing Dataset

```python
df = pd.read_csv('../data/processed/cleaned_churn.csv')
display(df)
```

### Convert TotalCharges

```python
df['TotalCharges'] = pd.to_numeric(
    df['TotalCharges'],
    errors='coerce'
)
```

### Remove Nulls

```python
df.dropna(inplace=True)
```

### Convert Churn to Numeric

```python
df['Churn'] = df['Churn'].astype(str).str.strip()

df['Churn'] = df['Churn'].map({
    'Yes': 1,
    'No': 0
})
```

### Remove customerID

```python
df.drop('customerID', axis=1, inplace=True)
```

### Convert categorical columns

```python
df = pd.get_dummies(
    df,
    drop_first=True
)
```

### Convert boolean columns to integers

```python
for col in df.columns:
    
    if df[col].dtype == 'bool':
        df[col] = df[col].astype(int)
```

```python
df = df.apply(pd.to_numeric)
print(df.dtypes)
```

### Tenure Segmentation

```python
def tenure_group(tenure):
    if tenure <= 12:
        return 'New'
    elif tenure <= 24:
        return 'Growing'
    elif tenure <= 48:
        return 'Loyal'
    return 'Long-Term'

df['tenure_group'] = df['tenure'].apply(tenure_group)
```

### High Value Customer

```python
average_charges = df['MonthlyCharges'].mean()
df['high_value_customer'] = np.where(df['MonthlyCharges'] > average_charges, 1, 0)
```

###  One Hot Encoding

```python
df_encoded = pd.get_dummies(df, drop_first=True)
```

### Save Engineered Dataset

```python
df_encoded.to_csv('../data/processed/engineered_churn.csv', index=False)
print(df_encoded.head())
```
